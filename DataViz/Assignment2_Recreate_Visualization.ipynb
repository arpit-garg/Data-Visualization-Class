```python
import csv
import matplotlib.pyplot as plt
from matplotlib import figure
```

```python
f = open("Fortune500dataset.csv")
csv_reader = csv.reader(f)
header = next(csv_reader)
```

```python
header
```

```python
data = {}
for name in header:
    data[name] = []
data
```

```python
for row in csv_reader:
    for name, value in zip(header, row):
        data[name].append(value)
```

```python
data
```

```python
data.keys()
```

```python
%matplotlib inline
```

```python
sector = data['Sector']
```

```python
sector
```

```python
plt.plot(data['Profit'], data['Revenue'], 'og')
```

```python
plt.bar(data['Revenue'], data['Profit'])
```

```python
import numpy as np
```

```python
profits = np.array(data['Profit'], dtype=int)
revenue = np.array(data['Revenue'], dtype=int)
employees = np.array(data['Employees'], dtype=int)
```

```python
profit_per_employee = np.true_divide(profits, employees)*1000
revenue_per_employee = np.true_divide(revenue, employees)*1000
```

```python
profit_per_employee
```

```python
revenue_per_employee
```

```python
N = len(profits)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
rects1 = ax1.bar(ind, revenue_per_employee, width, color='g')

rects2 = ax2.bar(ind + width, profit_per_employee, width, color='orange')

# add some text for labels, title and axes ticks
ax1.set_ylabel('Revenue per employee')
ax2.set_ylabel('Profit per employee')
ax1.set_title('Profits and Revenues per Employees for each sector')
ax1.set_xticks(ind + width / 2)
ax1.set_xticklabels(sector)
labels = ax1.get_xticklabels()
plt.setp(labels, rotation=40, fontsize=8)
ax1.legend((rects1[0], rects2[0]), ('Profit Per Employee', 'Revenue Per Employee'))
plt.show()
```
